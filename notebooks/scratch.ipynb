# webridge — scratch

Quick usage examples. Assumes `uv pip install -e ".[dynamic,discover,cli]"` and `playwright install chromium`.

```python
from webridge import fetch, search
```

## Fetch a page (static path, with automatic dynamic fallback)

```python
page = fetch("https://example.com/")
print(page.fetch_method, page.char_count)
print(page.markdown[:500])
```

## Force the dynamic path

```python
page = fetch("https://developer.samsung.com/health/sensor", method="dynamic")
print(page.title)
print(page.markdown[:500])
```

## Search

```python
for r in search("Polar BLE SDK", limit=5):
    print(r.rank, r.title)
    print(" ", r.url)
```
